```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#assign file path and import excel file from var 'Stormofswords'
Stormofswords = r'C:\Users\rjoga\Google Drive\People Analytics\ONA\00_DATA\stormofswords.xlsx'
GoT = pd.read_excel(Stormofswords)
```

```python
#Goal: Create some attribute data for the characters (e.g. House,Alive/dead,Age) to add to my edge_list
#Return to this once you've plotted your edgelist
Attribute_dat = GoT
```

```python
#Check columns to get parameters needed for networkx function
GoT.columns
```

```python
#create graph
network_graph = nx.from_pandas_edgelist(GoT,source='Source',target = 'Target',edge_attr = 'Weight')
```

```python
btw_centrality = nx.communicability_betweenness_centrality(network_graph)
d= network_graph.degree()
```

```python
#visualize graph
plt.figure(figsize=(60,60))
limit=plt.axis('off')
nx.draw_networkx(network_graph,
                 arrows=True,
                 node_color='b',node_size=d)
```
